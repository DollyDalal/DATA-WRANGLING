```python
#preprocessing on a dataset 1
import pandas as pd

data={
    "roll no":[101,102,103,104,105],
    "name":["asha","rohan","meera","karan","anil"],
    "age":[20,"-",21,20,"22"],
    "marks":[85,90,"-",88,76],
    "department":["CSE","IT","CSE","IT","ENTC"],
    "Admission date":["01-06-2023","03-06-2023","05-06-2023","07-06-2023","10-06-23"]
}
df=pd.DataFrame(data)
#save as CSV
df.to_csv("data.csv",index=False)
df

```

```python
#step 2: replace '-' with NaN to mark missing values
df.replace('-', np.nan, inplace=True)

#convert columns to numeric where needed
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df["marks"]=pd.to_numeric(df["marks"],errors='coerce')

'''when converting data(like strings- datetime),
if some values are invalid or cannot be converted,
instead of throwing an error, it will:
- replace those invalid values with NaT (Not a Time)'''

df

```

```python
#step 3: check missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
#fill missing age with mean
df['age'].fillna(df['age'].mean(), inplace=True)
#fill missing marks with median
df['marks'].fillna(df['marks'].median(), inplace=True)
df
```

```python
df
```

```python
#step 4
print(df.isnull().sum())
df
```

```python
#step 5: fix inconsistent text formatting
df["department"]=df["department"].str.upper()
df
```

```python
# Step 6: Convert Admission_date to standard datetime format

df["Admission date"] = pd.to_datetime(df["Admission date"], errors='coerce')

# display dataframe
df
```

```python
#step 7: save cleaned dataset
df.to_csv("cleaned_student_data.csv", index=False)
print("\nCleaned CSV file saved as cleaned_student_data")
```

IMPUTING THE MISSING VALUE

```python
#WORKING WITH MISSING VALUES ON DATASET-2
df=pd.read_csv('/content/Cars93.csv')
df
```

```python
#imputing the missing value
#1. Luggage room
df['Luggage.room'].fillna(df['Luggage.room'].mean(),inplace=True)
df
```

```python
df.isna().sum()
```

```python
#rear.seat.room
df['Rear.seat.room']=df['Rear.seat.room'].fillna(value=df['Rear.seat.room'].mean())
df
```

```python
#air bags
df['AirBags']=df['AirBags'].fillna(value=df['AirBags'].mode().values[0])
df
```

```python
#save updated dataset to csv
df.to_csv("updated_Cars93.csv", index=False)
print("\nUpdated CSV file saved as updated_Cars93 ")
```
